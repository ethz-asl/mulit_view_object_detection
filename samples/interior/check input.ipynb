```python
import interior_seq
import os
from  mrcnn import model_h as modellib
import sys
ROOT_DIR = os.path.abspath('../../')
sys.path.append(ROOT_DIR)  # To find local version of the library
from mrcnn import utils
import numpy as np
import matplotlib.pyplot as plt



NYU40_to_sel_map = {0: 0, 1: 0, 2: 0, 3: 1, 4: 2, 5: 3, 6: 4, 7: 5, 8: 0, 9: 0, 10: 8, 11: 6, 12: 0, 13: 7, 14: 5, 15: 8, 16: 0, 17: 9, 18: 10, 19: 11, 20: 0, 21: 12, 22: 0, 23: 13, 24: 14, 25: 15, 26: 16, 27: 17, 28: 0, 29: 0, 30: 0, 31: 0, 32: 0, 33: 18, 34: 19, 35: 20, 36: 21, 37: 22, 38: 0, 39: 0, 40: 0}
selected_classes = ['BG', 'cabinet', 'bed', 'chair', 'sofa', 'table', 
                'picture', 'blinds', 'shelves', 'dresser', 'pillow', 
                'mirror',  'clothes','books', 'refrigerator','television', 'paper', 'towel',
                'toilet', 'sink', 'lamp', 'bathtub', 'bag']
selected_class_list = [0, 3, 4, 5, 6, 7, 11, 13, 15, 17, 18, 19, 21, 23, 25, 26, 27, 33, 34, 35, 36, 37, 14, 10, 24]
# Configurations

dataset_dir = os.path.join(ROOT_DIR, 'data/InteriorNet/data/HD7')
dataset = interior_seq.InteriorDataset()
dataset.load_Interior(dataset_dir=dataset_dir, subset='train', class_ids=selected_class_list, 
                            NYU40_to_sel_map=NYU40_to_sel_map, selected_classes=selected_classes)
dataset.prepare()

class TrainConfig(interior_seq.InteriorNetConfig):
                TOP_DOWN_PYRAMID_SIZE = 24
                FPN_CLASSIF_FC_LAYERS_SIZE = 128
                POST_NMS_ROIS_INFERENCE = 1000
                POST_NMS_ROIS_TRAINING = 500
                PRE_NMS_LIMIT = 1500
                GPU_COUNT = 1
                IMAGES_PER_GPU = 2
                STEPS_PER_EPOCH = 5000
                VALIDATION_STEPS = 800
                NUM_CLASSES = len(selected_classes)  # background + num classes
                vmin = -1.07
                vmax = 1.07
                nvox = 40
                nvox_z = 40
                vsize = float(vmax - vmin) / nvox
                vox_bs = 1
                im_bs = 1
                samples = 10
                NUM_VIEWS = 2
                RECURRENT = False
                USE_RPN_ROIS = True
                LEARNING_RATE = 0.001
                GRID_REAS = 'ident'
                BACKBONE = 'resnet50'
config = TrainConfig()
image_ids = dataset.image_ids[:50]
min_mean = 50
for im_num, image_id in enumerate(image_ids):
    print("processing image {} of {}".format(im_num, len(image_ids))) 
    image, image_meta, gt_class_id, gt_bbox, gt_mask =\
        modellib.load_image_gt(dataset, config,
                               image_id, use_mini_mask=False)
#     image = dataset.load_image(image_id)
#     image, _, _, _, _ = utils.resize_image(
#                         image,
#                         min_dim=640,
#                         min_scale=2,
#                         max_dim=640,
#                         mode='square')
    
    if np.any(np.isnan(image)):
        print("image {} is nan".format(image_id))
    im_mean = image.mean()
    if im_mean < min_mean:
        min_mean = im_mean
        plt.imshow(image)
        print("new min:{}".format(im_mean))
```
